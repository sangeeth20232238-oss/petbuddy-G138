```python
# Column name that contains symptoms
SYMPTOM_COL = "symptom"

# Normalize symptom text (VERY IMPORTANT)
df[SYMPTOM_COL] = df[SYMPTOM_COL].astype(str).str.strip().str.lower()

col_order = ["case_id", "animalname", "symptom", "first_aid_advice", "emergency"]

# Keep any other columns (if you have) after these
rest = [c for c in df.columns if c not in col_order]
df = df[col_order + rest]

# --- Create empty columns first ---
if "first_aid_advice" not in df.columns:
    df.insert(len(df.columns), "first_aid_advice",
              "This symptom requires veterinary assessment. Please contact a veterinarian.")

if "emergency" not in df.columns:
    df.insert(len(df.columns), "emergency", "no")

# --- Fill columns where advice exists ---
for symptom, info in FIRST_AID_MAP.items():
    mask = df[SYMPTOM_COL] == symptom
    df.loc[mask, "first_aid_advice"] = info["advice"]
    df.loc[mask, "emergency"] = info["emergency"]          # yes / no
```
