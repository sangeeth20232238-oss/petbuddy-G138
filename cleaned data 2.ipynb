```python
data2["Animal_type"]=data2["Animal_Type"].astype(str).str.strip().str.lower()
```

```python
print(data2.head())
```

```python
cols_to_remove = [
    "Breed","Age", "Gender", "Weight", "Duration", "Appetite_Loss", "Vomiting",
    "Diarrhea", "Coughing", "Labored_Breathing", "Lameness", "Skin_Lesions",
    "Nasal_Discharge", "Eye_Discharge", "Body_Temperature", "Heart_Rate",
    "Disease_Prediction"
]

data2 = data2.drop(columns=cols_to_remove, errors='ignore')
```

```python
data2 = data2.rename(columns={"Animal_Type": "AnimalName"})
```
